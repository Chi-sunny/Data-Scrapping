```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://jiji.ng/computers-and-laptops"
```

```python
r = requests.get(url)
```

```python
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
print(soup)
```

```python
names = soup.find_all("div", class_="b-advert-title-inner" )
```

```python
print(names)
```

```python
product_names = []
```

```python
for i in names:
    name = i.text
    product_names.append(name)
```

```python
print(product_names)
```

```python
prices = soup.find_all("div", class_="qa-advert-price")
```

```python
prices_list = []
for i in prices:
    price = i.text
    prices_list.append(price)
```

```python
print(prices_list)
```

```python
list_of_prices = [x.text for x in prices]
list_of_names = [x.text for x in names]
```

```python
print(list_of_names)
```

```python
print(list_of_prices)
```

```python
df = pd.DataFrame({"Product Names": product_names, "Prices": prices_list})
```
